Based on the derivations outlined in the paper, here are the explicit mathematical steps to transition from **Equation (3)** to **Equation (4)**.

[cite_start]The paper states that one must expand $|\Psi_0(\lambda+\epsilon)\rangle$ for small $\epsilon$[cite: 88]. [cite_start]Since Equation (4) is valid for **unnormalized** wavefunctions[cite: 91], we must account for the normalization factor in the definition of fidelity during the expansion.

### 1. Starting Equations
* **Fidelity Susceptibility (Eq. 3):**
    $$\chi_F(\lambda) = - \frac{\partial^2 \ln F}{\partial \epsilon^2} \bigg|_{\epsilon=0}$$
* **Fidelity (Eq. 2):**
    $$F(\lambda, \epsilon) = |\langle\Psi_0(\lambda)|\Psi_0(\lambda+\epsilon)\rangle|$$
    *Note: To derive the general form Eq. (4), we treat the fidelity as the normalized overlap:*
    $$F = \frac{|\langle \Psi_0(\lambda) | \Psi_0(\lambda+\epsilon) \rangle|}{\sqrt{\langle \Psi_0(\lambda) | \Psi_0(\lambda) \rangle \langle \Psi_0(\lambda+\epsilon) | \Psi_0(\lambda+\epsilon) \rangle}}$$

### 2. Taylor Expansion
Let $|\Psi\rangle = |\Psi_0(\lambda)\rangle$ and $|\Psi_\epsilon\rangle = |\Psi_0(\lambda+\epsilon)\rangle$.
We perform a Taylor expansion of the perturbed state up to second order in $\epsilon$:
$$|\Psi_\epsilon\rangle \approx |\Psi\rangle + \epsilon |\partial_\lambda \Psi\rangle + \frac{\epsilon^2}{2} |\partial^2_\lambda \Psi\rangle$$

### 3. Expanding the Log-Fidelity terms
We calculate $\ln F = \frac{1}{2} \ln |\langle \Psi | \Psi_\epsilon \rangle|^2 - \frac{1}{2} \ln \langle \Psi_\epsilon | \Psi_\epsilon \rangle - \frac{1}{2} \ln \langle \Psi | \Psi \rangle$. We focus on the terms dependent on $\epsilon$.

**A. Expand the Overlap Term** ($\ln |\langle \Psi | \Psi_\epsilon \rangle|^2$):
First, expand the inner product:
$$\langle \Psi | \Psi_\epsilon \rangle = \langle \Psi | \Psi \rangle + \epsilon \langle \Psi | \partial_\lambda \Psi \rangle + \frac{\epsilon^2}{2} \langle \Psi | \partial^2_\lambda \Psi \rangle$$
The squared modulus is $|\langle \Psi | \Psi_\epsilon \rangle|^2 = \langle \Psi | \Psi_\epsilon \rangle \langle \Psi_\epsilon | \Psi \rangle$. Keeping terms up to $\epsilon^2$:
$$\approx \langle \Psi | \Psi \rangle^2 \left[ 1 + \epsilon \frac{\langle \Psi | \partial \Psi \rangle + \langle \partial \Psi | \Psi \rangle}{\langle \Psi | \Psi \rangle} + \epsilon^2 \left( \frac{\langle \Psi | \partial \Psi \rangle \langle \partial \Psi | \Psi \rangle}{\langle \Psi | \Psi \rangle^2} + \frac{\langle \Psi | \partial^2 \Psi \rangle + \langle \partial^2 \Psi | \Psi \rangle}{2\langle \Psi | \Psi \rangle} \right) \right]$$
Using $\ln(1+x) \approx x - x^2/2$, the log of this term is:
$$\ln |\langle \Psi | \Psi_\epsilon \rangle|^2 \approx \text{const} + \epsilon (\dots) + \epsilon^2 \left[ \frac{\langle \Psi | \partial \Psi \rangle \langle \partial \Psi | \Psi \rangle}{\langle \Psi | \Psi \rangle^2} + \frac{\text{Re}\langle \Psi | \partial^2 \Psi \rangle}{\langle \Psi | \Psi \rangle} - \frac{1}{2} \left( \frac{2\text{Re}\langle \Psi | \partial \Psi \rangle}{\langle \Psi | \Psi \rangle} \right)^2 \right]$$

**B. Expand the Normalization Term** ($\ln \langle \Psi_\epsilon | \Psi_\epsilon \rangle$):
$$\langle \Psi_\epsilon | \Psi_\epsilon \rangle \approx \langle \Psi | \Psi \rangle + \epsilon(\langle \Psi | \partial \Psi \rangle + \langle \partial \Psi | \Psi \rangle) + \epsilon^2 (\langle \partial \Psi | \partial \Psi \rangle + \frac{1}{2}\langle \Psi | \partial^2 \Psi \rangle + \frac{1}{2}\langle \partial^2 \Psi | \Psi \rangle)$$
Taking the log $\ln(1+x) \approx x - x^2/2$:
$$\ln \langle \Psi_\epsilon | \Psi_\epsilon \rangle \approx \text{const} + \epsilon (\dots) + \epsilon^2 \left[ \frac{\langle \partial \Psi | \partial \Psi \rangle + \text{Re}\langle \Psi | \partial^2 \Psi \rangle}{\langle \Psi | \Psi \rangle} - \frac{1}{2} \left( \frac{2\text{Re}\langle \Psi | \partial \Psi \rangle}{\langle \Psi | \Psi \rangle} \right)^2 \right]$$

### 4. Constructing the Susceptibility
We define $\chi_F$ using the second derivative: $\chi_F = - \frac{\partial^2}{\partial \epsilon^2} \left( \frac{1}{2} \ln |\langle \Psi | \Psi_\epsilon \rangle|^2 - \frac{1}{2} \ln \langle \Psi_\epsilon | \Psi_\epsilon \rangle \right) \bigg|_{\epsilon=0}$.

When we subtract term **B** from term **A**:
1.  The **second derivative terms** $\langle \Psi | \partial^2 \Psi \rangle$ cancel out.
2.  The **squared linear terms** (arising from the Taylor expansion of $\ln$) cancel out.
3.  The remaining terms at order $\epsilon^2$ are:
    $$\frac{1}{2} \left( \frac{\langle \partial \Psi | \Psi \rangle \langle \Psi | \partial \Psi \rangle}{\langle \Psi | \Psi \rangle^2} - \frac{\langle \partial \Psi | \partial \Psi \rangle}{\langle \Psi | \Psi \rangle} \right) \epsilon^2$$

### 5. Final Result
Taking the derivative $-\frac{\partial^2}{\partial \epsilon^2}$ removes the $\frac{1}{2}\epsilon^2$ factor and flips the sign:

$$\chi_F(\lambda) = \frac{\langle \partial_\lambda \Psi_0 | \partial_\lambda \Psi_0 \rangle}{\langle \Psi_0 | \Psi_0 \rangle} - \frac{\langle \Psi_0 | \partial_\lambda \Psi_0 \rangle}{\langle \Psi_0 | \Psi_0 \rangle} \frac{\langle \partial_\lambda \Psi_0 | \Psi_0 \rangle}{\langle \Psi_0 | \Psi_0 \rangle}$$

[cite_start]This matches **Equation (4)** exactly[cite: 89].

```python
from functools import reduce
from scipy.sparse.linalg import eigsh
from functools import lru_cache

# Pauli matrices
I = np.eye(2)
X = np.array([[0, 1], [1, 0]])
Y = np.array([[0, -1j], [1j, 0]])
Z = np.array([[1, 0], [0, -1]])

def kron_n(ops):
    return reduce(np.kron, ops)

@lru_cache(maxsize=None)
def tfim_hamiltonian(n, J=1.0, h=1.0, eps=0, pbc=False):
    # \mathcal{H}_{\text{TFIM}} = -J \sum_{\langle i,j \rangle} \sigma_i^z \sigma_j^z - h \sum_i \sigma_i^x
    H = np.zeros((2**n, 2**n), dtype=complex)
    for i in range(n - 1):
        ops = [I]*n
        ops[i], ops[i+1] = Z, Z
        H -= J * kron_n(ops)
    if pbc:
        ops = [I]*n
        ops[0], ops[-1] = Z, Z
        H -= J * kron_n(ops)
        
    for i in range(n):
        ops = [I]*n
        ops[i] = X
        H -= h * kron_n(ops)

        # add small epsilon bias to break degeneracy
        ops = [I]*n
        ops[i] = Z
        H -= eps * kron_n(ops)

    return H


# get ground state via lanczos
def ground_state(H, use_lanczos=False):
    if use_lanczos:
        vals, vecs = eigsh(H, k=1, which='SR')  # using lanczos to get smallest eigenvalue
    else:
        vals, vecs = np.linalg.eigh(H)  # full diagonalization

    
    # GAUGE FIXING:
    # Ensure the first non-zero element is real and positive 
    # to align phase across different h steps.
    first_element = vecs[0, 0]
    if np.abs(first_element) > 1e-10:
        phase_factor = np.abs(first_element) / first_element
        vecs[:, 0] = vecs[:, 0] * phase_factor
    return vals[0], vecs[:, 0]
```

```python
n=8
H = tfim_hamiltonian(n, eps=1e-6, pbc=False)
# U = marshall_U(n)
# H = U @ H @ U  # transform to stoquastic form
E0, psi0 = ground_state(H)
# E0, psi0 = ground_state(H)
print("Ground state energy:", E0)

E0_, psi0_ = ground_state(H, use_lanczos=True)
psi0_m = psi0_.copy()
# set imag part to zero
psi0_m.imag = 0.0

print("Ground state energy (lanczos):", E0_)
np.linalg.norm(H@psi0_/E0_ -psi0_)  # should be small
print("overlap between lanczos and full diag:", np.abs(np.vdot(psi0_m, psi0)))
print("energy difference:", np.abs(E0 - E0_))
```

```python
@lru_cache(maxsize=None)
def ground_state_wrapper(h, n, J=1.0, pbc=False, use_lanczos=True):
    H = tfim_hamiltonian(n, J=J, h=h, eps=1e-6, pbc=pbc)
    _, gs = ground_state(H, use_lanczos=use_lanczos)

        
    return gs

@lru_cache(maxsize=None)
def dpsi_dh(h0, n, delta_h=1e-5, J=1.0, pbc=False, use_lanczos=True):
    psi_plus = ground_state_wrapper(h0 + delta_h, n, J=J, pbc=pbc, use_lanczos=use_lanczos)
    psi_minus = ground_state_wrapper(h0 - delta_h, n, J=J, pbc=pbc, use_lanczos=use_lanczos)
    dpsi = (psi_plus - psi_minus) / (2 * delta_h)
    return dpsi

def chi_F(h0, n, delta_h=1e-5, J=1.0, pbc=False, use_lanczos=True):
    dpsi = dpsi_dh(h0, n, delta_h=delta_h, J=J, pbc=pbc, use_lanczos=use_lanczos)
    psi0 = ground_state_wrapper(h0, n, J=J, pbc=pbc, use_lanczos=use_lanczos)

    # chi = <dpsi | dpsi>/<psi0|psi0> - |<psi0 | dpsi>|^2/<psi0|psi0>
    N =np.vdot(psi0, psi0)
    chi = (np.vdot(dpsi, dpsi) / N) - (np.abs(np.vdot(psi0, dpsi))**2 / N**2)
    return chi
```

```python
all_chi = []
hs = np.linspace(0.5, 1.5, 30)
n = 8
for h in tqdm(hs):
    chi = chi_F(h, n, J=1.0, pbc=True, use_lanczos=True, delta_h=1e-5)
    all_chi.append(chi) 
    print(f"Fidelity susceptibility at h={h}: {chi.real}")
```


## 2. Larger systems: Monte Carlo estimate

For larger $N$ you cannot sum over all configurations, so you estimate
$$
\langle \psi_A|\psi_B\rangle
= \sum_{\boldsymbol{\sigma}}
|\psi_A(\boldsymbol{\sigma})|^2\,
\frac{\psi_B(\boldsymbol{\sigma})}{\psi_A(\boldsymbol{\sigma})}
$$
by sampling $\boldsymbol{\sigma}$ from
$p_A(\boldsymbol{\sigma}) \propto |\psi_A(\boldsymbol{\sigma})|^2$
(which you already do in VMC).

### Two RBMs

Assume both states are (approximately) normalized and you can draw samples
$\boldsymbol{\sigma}_k$ from $|\psi_A|^2$.

1. Use your existing Markov chain to generate
   $\boldsymbol{\sigma}_k\sim |\psi_A|^2$, for $k=1,\dots,M$.
2. For each sample compute the ratio
   $$
   r_k = \frac{\psi_B(\boldsymbol{\sigma}_k)}{\psi_A(\boldsymbol{\sigma}_k)}
   $$
   via log-amplitudes:
   $$
   \log r_k = \log\psi_B(\boldsymbol{\sigma}_k) - \log\psi_A(\boldsymbol{\sigma}_k).
   $$
3. Estimator for the overlap of normalized states:
   $$
   \langle \psi_A|\psi_B\rangle \approx \frac{1}{M}\sum_{k=1}^M r_k.
   $$
4. The fidelity is then
   $$
   F \approx \left|\frac{1}{M}\sum_{k=1}^M r_k\right|^2.
   $$

If your RBMs are not exactly normalized you can either explicitly normalize
them (estimate $\|\psi\|$ with a separate Monte Carlo or fix the convention
during training) or work with this as an approximation, which is standard
in NQS practice.

### RBM vs exact groundstate

Assume the exact state is normalized and you can either

* sample configurations $\boldsymbol{\sigma}$ from $|\psi_{\text{exact}}|^2$
  (if you have a sampler), or
* sample uniformly over a subset when $N$ is not too large.

If you can sample from $|\psi_{\text{exact}}|^2$:

1. Draw $\boldsymbol{\sigma}_k \sim |\psi_{\text{exact}}|^2$.
2. Compute
   $$
   r_k = \frac{\psi_{\text{RBM}}(\boldsymbol{\sigma}_k)}
              {\psi_{\text{exact}}(\boldsymbol{\sigma}_k)}.
   $$
3. Estimator:
   $$
   \langle \psi_{\text{RBM}}|\psi_{\text{exact}}\rangle
   \approx \frac{1}{M}\sum_{k=1}^M r_k.
   $$

If you only have the full exact vector and $N$ is too large to enumerate,
choose some easier proposal distribution $q(\boldsymbol{\sigma})$
(e.g. product state, or a truncated exact distribution), then use importance
sampling:
$$
\langle \psi_{\text{RBM}}|\psi_{\text{exact}}\rangle
= \mathbb{E}_{\boldsymbol{\sigma}\sim q}\Big[
\frac{\psi_{\text{RBM}}^*(\boldsymbol{\sigma})\psi_{\text{exact}}(\boldsymbol{\sigma})}
     {q(\boldsymbol{\sigma})}
\Big]
$$
and estimate that with samples.


You just need the usual Hilbert-space overlap
$$
\langle \psi_A|\psi_B\rangle
= \sum_{\boldsymbol{\sigma}} \psi_A^*(\boldsymbol{\sigma}) \psi_B(\boldsymbol{\sigma})
$$

An RBM wavefunction:
$$
\psi_{\text{RBM}}(\boldsymbol{\sigma})
= \exp\Big(\sum_i a_i \sigma_i\Big)
\prod_j 2\cosh\Big(b_j + \sum_i W_{ij}\sigma_i\Big)
$$
